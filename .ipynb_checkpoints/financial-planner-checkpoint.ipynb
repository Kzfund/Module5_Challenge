## Part 1 - Personal Finance Planner

### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
# YOUR CODE HERE!
my_btc = 1.2
my_eth = 5.3

```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Fetch current BTC price
# YOUR CODE HERE!
btc_response_data = requests.get(btc_url).json()
btc_response_data
# Fetch current ETH price
# YOUR CODE HERE!
eth_response_data = requests.get(eth_url).json()
eth_response_data
# Compute current value of my crpto
# YOUR CODE HERE!
btc_current_value = my_btc * btc_response_data['data']['1']['quotes']['CAD']['price']
eth_current_value = my_eth * eth_response_data['data']['1027']['quotes']['CAD']['price']
# Print current crypto wallet balance
print(f"The current value of your {my_btc} BTC is ${btc_current_value:0.2f}")
print(f"The current value of your {my_eth} = ETH is ${eth_current_value:0.2f}")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
alpaca_end_point = os.getenv("ALPACA_END_POINT")
# YOUR CODE HERE!
# Create the Alpaca API object
# YOUR CODE HERE!

alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2",
    base_url= alpaca_end_point)
```

```python
# Format current date as ISO format
# YOUR CODE HERE!
today = pd.Timestamp("2023-09-27", tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to "1Day" for Alpaca API
timeframe = "1Day"

# Get current closing prices for SPY and AGG
# YOUR CODE HERE!
df_stock_data = alpaca.get_bars(
    tickers,
    timeframe,
    start = today,
    end = today,
).df


df_stock_data

# Reorganize the DataFrame
# Separate ticker data
# YOUR CODE HERE!
AGG = df_stock_data[df_stock_data['symbol']=='AGG'].drop('symbol', axis=1)
SPY = df_stock_data[df_stock_data['symbol']=='SPY'].drop('symbol', axis=1)

# Concatenate the ticker DataFrames
# YOUR CODE HERE!
df_stock_data = pd.concat([AGG, SPY],axis=1, keys=['AGG','SPY'])

# Preview DataFrame
# YOUR CODE HERE!
df_stock_data
```

```python
# Pick AGG and SPY close prices
# YOUR CODE HERE!
agg_close_price = float(df_stock_data["AGG"]["close"][0])
spy_close_price = float(df_stock_data["SPY"]["close"][0])


# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
# YOUR CODE HERE!
my_spy_value = my_spy * spy_close_price
my_agg_value = my_agg * agg_close_price
# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
# YOUR CODE HERE!
monthly_income = 12000
# Consolidate financial assets data
# YOUR CODE HERE!
crypto_assets = btc_current_value + eth_current_value
stock_bond_assets = my_spy_value + my_agg_value

value_data = {
    'amount': [crypto_assets, stock_bond_assets],
    'asset': ["crypto", "shares"]
}
# Create savings DataFrame
# YOUR CODE HERE!
df_savings = pd.DataFrame(value_data).set_index('asset')
# Display savings DataFrame
display(df_savings)
```

```python
# Plot savings pie chart
# YOUR CODE HERE!
pie_chart = df_savings.plot.pie(y= 'amount', title="Savings Portfolio Composition", figsize=(5,5))
```

```python
# Set ideal emergency fund
emergency_fund = monthly_income * 3

# Calculate total amount of savings
# YOUR CODE HERE!
total_savings = df_savings['amount'].sum()
# Validate saving health
# YOUR CODE HERE!
if total_savings > emergency_fund:
    print(f"Congratulation, you have enough money in your emergency fund.")
elif total_savings == emergency_fund:
    print(f"You have reached your financial goal.")
else:
    print(f"You are ${emergency_fund - total_savuings:0.2f} away from achieving your financial goal.")
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of thirty years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date_30 = pd.Timestamp('1993-09-22', tz='America/New_York').isoformat()
end_date_30 = pd.Timestamp('2023-09-22', tz='America/New_York').isoformat()
```

```python
#Set Tickers
tickers = ['AGG', 'SPY']
```

```python
# Set timeframe to "1Day" for Alpaca API
timeframe = "1Day"

# # Get 30 years' worth of historical data for SPY and AGG
df_ticker_30 = alpaca.get_bars(
    tickers,
    timeframe,
    start=start_date_30,
    end=end_date_30
).df
```

```python
df_ticker_30.tail()
```

```python
# Reorganize the DataFrame
# Separate ticker data
spy_30 = df_ticker_30[df_ticker_30['symbol']=='SPY'].drop('symbol', axis=1)
agg_30 = df_ticker_30[df_ticker_30['symbol']=='AGG'].drop('symbol', axis=1)

# Concatenate the ticker DataFrames
df_30 = pd.concat([spy_30, agg_30], axis=1, keys=["SPY", "AGG"])

# Display sample data
df_30.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
MC_thirtyyear = MCSimulation(
    portfolio_data = df_30,
    weights = [.60,.40],
    num_simulation = 500,
    num_trading_days = 252*5
)
```

```python
# Printing the simulation input data
MC_thirtyyear.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!
MC_thirtyyear.calc_cumulative_return()
```

```python
# Plot simulation outcomes
# YOUR CODE HERE!
line_plot = MC_thirtyyear.plot_simulation()
#save plot
line_plot.get_figure().savefig("MC_thirtyyear_sim_plot.png", bbox_inches="tight")
```

```python
# Plot probability distribution and confidence intervals
# YOUR CODE HERE!
dist_plot = MC_thirtyyear.plot_distribution()
#save plot
dist_plot.get_figure().savefig('MC_thirtyyear_dist_plot.png',bbox_inches='tight')
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
# YOUR CODE HERE!
tbl = MC_thirtyyear.summarize_cumulative_return()

# Print summary statistics
# YOUR CODE HERE!
print(tbl)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
# YOUR CODE HERE!
ci_lower = round(tbl[8]*initial_investment,2)
ci_upper = round(tbl[9]*initial_investment,2)
# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
# YOUR CODE HERE!
ci_lower = round(tbl[8]*initial_investment,2)
ci_upper = round(tbl[9]*initial_investment,2)
# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```
